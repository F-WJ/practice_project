```python
import tushare as ts
import config

# 设置token
ts.set_token(config.mytoken)
# 初始化pro接口
pro = ts.pro_api()
# 数据调取
df = pro.bo_daily(date='20181102')
print(df)
type(df)
df.head(10)
```

# pyecharts可视化

```python
# 单日调查
from pyecharts import Pie

moviename = df.name
movieamount = df.day_amount
pie = Pie("单日票房", title_pos='center', width=900)
pie.add("票房", moviename, movieamount, center=[45, 60], is_random=True, radius=[30,75], rosetype='area', 
        is_label_show=True, is_legend_show=False)
pie
```

```python
# 一年总票房
import pandas as pd

number = 11
years = '2018'
months = ['02', '03', '04', '05', '06', '07', '08', '09']
January = years + '01' + '01'
df_jan = pro.bo_monthly(date=January)

for month in months:
        date = years + month + '01'
        df_month = pro.bo_monthly(date=date)
        for a in range(10):
            for b in range(10):
                if df_month.name[a] == df_jan.name[b]:
                    result = True
            if result:
                df_jan.loc[number] = df_month.loc[a]
                number = number + 1
#                 print(df_jan.head(50))
                
        for x in range(10):
            for y in range(10):
                if df_jan.name[x] == df_month.name[y]:
                    df_jan.month_amount[x] = df_jan.month_amount[x] + df_month.month_amount[y]
                
                
                
        # 删除不必要的列
to_drop = ['list_date','avg_price','list_day','m_ratio','p_pc','rank', 'wom_index']
df_jan.drop(to_drop, inplace=True, axis=1)
# df_jan.head(200)



from pyecharts import Pie

moviename = df_jan.name
movieamount = df_jan.month_amount
pie = Pie("全年票房", title_pos='center', width=1000, height= 800)
pie.add("票房", moviename, movieamount, center=[50, 50], is_random=True, radius=[30,50], rosetype='area', 
        is_label_show=True, is_legend_show=False)
pie
```
